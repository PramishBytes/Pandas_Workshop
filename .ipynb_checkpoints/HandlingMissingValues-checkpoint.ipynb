## Fillna

__Definition and Usage__

The __fillna()__ method replaces the NULL values with a specified value.

The __fillna()__ method returns a new DataFrame object unless the __inplace__ parameter is set to __True__, in that case the __fillna()__ method does the replacing in the original DataFrame instead 

__SYNTAX__: 

__dataframe.fillna(value,method,axis,inplace,limit,downcast)__

```python
var.fillna("Empty")
```

## Frontfill && Backwardfill

__Syntax__

__FrontFill__: dataframe.fillna(*,axis=None,inplace =False, limit=None,limi_area = None,downcast =<no_default>)


__BackFill__: dataframe.bfill(*,axis=None,inplace =False, limit=None,limi_area = None,downcast =<no_default>)

var.fillna(method - "ffill",axis =1)
## Or
var.fillna(method = "bfill",axis = 1)

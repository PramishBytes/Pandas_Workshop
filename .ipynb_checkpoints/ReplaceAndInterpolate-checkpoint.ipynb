## Replace & Interpolate

__SYNTAX:__

__DataFrame.replce(to_replace = None, vlaue = no_default,*,inplace = False, limit = None, regex = False, method = no_default)__

Replace values given in to_replace with value. 

Values of the Series/DataFrame are replaced with other values dynamically. This differ from updating with __.loc__ or __.iloc__, which require you to specify a location to update with some value.
